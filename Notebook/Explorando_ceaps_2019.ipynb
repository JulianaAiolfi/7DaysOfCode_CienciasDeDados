#Explorando dados Ceaps ano de 2019

##Limpeza e formatacao dos dados e salvando um novo csv

```python
import pandas as pd
```

```python
#Leitura DataFrame

dados = pd.read_csv('despesa_ceaps_2019.csv',encoding='latin1', delimiter=';',skiprows=1)
```

Vamos visualizar as cinco primeiras linhas e cinco últimas linhas do DataFrame

```python
#As cinco primeiras linhas dos DataFrame

dados.head()
```

```python
#As cinco últimas linhas dos DataFrame

dados.tail()
```

```python
#Informacoes sobre o DataFrame Dtype

dados.info()
```

Verificando os dados básicos estatísticos, podemos verificar que a coluna cod_documento esta com valores cientificos. A mesma nao deveria aparecer precisa ser alterada para objeto

```python
#Verificando dados básicos estatísticos

dados.describe()
```

Na terceira linha do DataFrame temos um dado referente a 2018, como estamos tratando somente 2019 precisamos excluir essa informacao

```python
#Excluindo a terceira linha com a informacao de 2018

dados = dados.drop(3)
```

A colunas 'Valor_reembolso' esta sendo formatada, alterando de (,) para (.) e o tipo da coluna de str para inteiro

A coluna Cod_documento formatamos de inteiro para objeto  

```python
#Formatando a coluna de valor_reembolso em inteiro e , para .

dados['VALOR_REEMBOLSADO'] = dados['VALOR_REEMBOLSADO'].str.replace(',', '.')
dados['VALOR_REEMBOLSADO'] = pd.to_numeric(dados['VALOR_REEMBOLSADO'])
dados['VALOR_REEMBOLSADO'] = dados['VALOR_REEMBOLSADO'].astype(int)

#Alterando a coluna cod_documento de inteiro para objeto

dados['COD_DOCUMENTO'] = dados['COD_DOCUMENTO'].astype(object)
```

```python
#Formatando a coluna de data em datetime, deixando datas invalidas como nulas

dados['DATA'] = pd.to_datetime(dados['DATA'], errors='coerce')
```

```python
#Visualizando as informacoes da coluna tipo_despesa

dados['TIPO_DESPESA'].unique()
```

```python
#Alterando as informacoes para valores objetivos

dados['TIPO_DESPESA'] = dados['TIPO_DESPESA'].map({'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.': 'Aluguel',
                                             'Divulgação da atividade parlamentar': 'Divulgação',
                                             'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ': 'Despesas gerais',
                                             'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens',
                                             'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar': 'Contratações',
                                             'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Locomoção, hospedagem e alimentação',
                                             'Serviços de Segurança Privada': 'Segurança'})
```

```python
#Visualizando os tipos alterado

dados.info()
```

```python
#Dados estatísticos, com valor_reembolso 

dados.describe()
```

```python
#Salvando um novo dataset com as informacoes alteradas

dados.to_csv('dados_limpos_ceaps_2019.csv', index=False)
```
